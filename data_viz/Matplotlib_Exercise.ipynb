The dataset given for this notebook is the pricing data for diamonds based on the carat, cut, colour and so on. Take a look at the data below and also the data dictionary for a better understanding of the data and the attributes.

```python
# Import the necessary libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read the data into a dataframe
data = pd.read_csv("diamonds.csv", header = 0)

# As a preparatory step, let's drop Unnamed column.
data = data.drop("Unnamed: 0", axis =1 )

data.head()
```

### Data Dictionary:

1. **carat**: Weight of the diamond (0.2--5.01)

2.  **cut**: Quality of the cut (Fair, Good, Very Good, Premium, Ideal)

3. **color**: Diamond colour, from J (worst) to D (best)

4. **clarity**: A measurement of how clear the diamond is (I1 (worst), SI2, SI1, VS2, VS1, VVS2, VVS1, IF (best))

5. **depth**: Total depth percentage = z / mean(x, y) = 2 * z / (x + y) (43--79)

6. **table**: Width of top of diamond relative to widest point (43--95)

7. **price**: Price in US dollars (326 dollars -18,823 dollars)

8. **x**: Length in mm (0--10.74)

9. **y**: Width in mm (0--58.9)

10. **z**: Depth in mm (0--31.8)

<h3 style = "color : Brown"> Question </h3>

Investigate the variation of prices of diamonds and note your observations.

```python
### The get the distribution of diamond prices
_=plt.hist(data.price,bins=100)
plt.xlabel('price of diamond')
plt.title('Distribution of Diamond price',fontdict={'fontsize':18,'fontweight':20,'color':'red'})
```

The distribution is right-skewed. As the price of the diamond increases, the number of diamonds on that category is very less. Expensive diamonds are very rare in quantity.

<h3 style = "color : Brown"> Question </h3>

Perform a similar variation analysis on the quality of cut.

```python
# (Fair, Good, Very Good, Premium, Ideal)
plt.boxplot([data[data['cut']=="Fair"].price,data[data['cut']=="Good"].price,data[data['cut']=="Very Good"].price,data[data['cut']=="Premium"].price,data[data['cut']=="Ideal"].price])
labels=("Fair", "Good", "Very Good", "Premium", "Ideal")
plt.xticks(range(1,len(labels)+1),labels)
plt.show()
```

```python
plt.bar(data.cut.value_counts().index,data.cut.value_counts().values)
plt.xticks(range(0,len(labels)),data.cut.unique())
plt.show()
```

Ideal cut diamonds are produced more in numbers and the price of them is skewed.

<h3 style = "color : Brown"> Question </h3>

Represent the same information as above in a pie chart.

```python
plt.pie(data.cut.value_counts().values,labels=data.cut.value_counts().index)
plt.show()
```

<h3 style = "color : Brown"> Question </h3>

Investigate the variation of price of the diamonds, with respect to the carats.

```python
plt.hist(data.carat,bins=100,range=[data.carat.min(),data.carat.max()])
plt.show()
plt.scatter(data.carat,data.price)
# sns.regplot(data.carat,data.price,label="fitted line")
plt.xlabel('carat')
plt.ylabel('price')
```

There's an increasing relationship seen between carat and price. The price of the diamond keeps increasing as the carat size increases.
